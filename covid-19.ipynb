```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Drop rows with missing values in 'TotalCases' or 'TotalDeaths'
df_cleaned = df.dropna(subset=['TotalCases', 'TotalDeaths'])
x = df_cleaned[['TotalCases']]
y = df_cleaned[['TotalDeaths']]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
predicted=model.predict(x)
print(predicted)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, predicted, color='red', linewidth=2, label='Linear Regression Fit')
plt.xlabel('Total Cases')
plt.ylabel('Total Deaths')
plt.title('Total Cases vs Total Deaths ')
plt.legend()
plt.grid(True)
plt.show()
```
